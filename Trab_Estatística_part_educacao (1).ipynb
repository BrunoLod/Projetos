**Buscando o teste de hipótese:**

```python
df
```

```python
from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
# Criando a variável que armazenará uma função que me permite
# testar se há diferenças significativas entre os grupos:

# A variável categórica nesse contexto é o nível de educação formal
# de cada grupo, sendo a variável quantitativa o salário.

result = f_oneway(df["Salary"][df["Education_Level"] == "High School"],
                  df["Salary"][df["Education_Level"] == "Bachelor's"],
                  df["Salary"][df["Education_Level"] == "Master's"],
                  df["Salary"][df["Education_Level"] == "PhD" ])
```

```python
# Passo 4: Analisar os Resultados do Teste de ANOVA
print("Resultado do teste ANOVA:")
print("Valor-p:", result.pvalue)
```

O teste de hipótese se consistia em dizer de que o salário não é influenciado pelo nível educacional (hiótese nula). Contudo, o teste do valor_p foi muito pequeno, próximo de zero, o que nos permite dizer que estasticamente, em termos gerais, há evidências significativas de que o nível de educação formal interfere no salário recebido.

```python
# Verificar se o valor-p é menor que o nível de significância (por exemplo, 0.05)
if result.pvalue < 0.05:
    print("Rejeitar a hipótese nula. Há evidências de diferença significativa nos salários.")

    # Passo 5: Realizar Testes Post Hoc (Teste de Tukey)
    tukey_results = pairwise_tukeyhsd(df['Salary'], df['Education_Level'])

    print("\nResultados do teste de Tukey:")
    print(tukey_results)

    # Visualizar os resultados do teste de Tukey
    tukey_results.plot_simultaneous(comparison_name='High School', figsize=(8, 4))
    plt.show()
else:
    print("Não há evidências suficientes para rejeitar a hipótese nula.")
```

**Calculando o intervalo de confiança:**

```python
import scipy.stats as stats
import statsmodels.api as sm
```

```python
high_school = df["Salary"][df["Education_Level"] == "High School"]
bachelor = df["Salary"][df["Education_Level"] == "Bachelor's"]
master = df["Salary"][df["Education_Level"] == "Master's"]
phd = df["Salary"][df["Education_Level"] == "PhD"]

# Calculando as médias e desvios padrãos para cada grupo:

means = [high_school.mean(), bachelor.mean(), master.mean(), phd.mean()]
std_devs = [high_school.std(), bachelor.std(), master.std(), phd.std()]

# Calculando o tamanho das amostras para cada grupo:

nobs = [len(high_school), len(bachelor), len(master), len(phd)]

# Calculando a média global e o desvio padrão global:

global_mean = np.mean(df["Salary"])
global_std_dev = np.std(df["Salary"])

# Calculando o erro padrão da diferença entre as médias:

standard_errors = [global_std_dev * np.sqrt(1 / nobs[i] + 1 / len(df["Salary"])) for i in range(4)]

# Calculando o intervalo de confiança:

conf_ints = [stats.t.interval(0.95, len(df["Salary"]) - 1, loc=means[i] - global_std_dev, scale=standard_errors[i]) for i in range(4)]

# Exibe os resultados
for i in range(4):
    print(f"Intervalo de Confiança para a diferença entre a média de {df['Education_Level'].unique()[i]} e a média global: {conf_ints[i]}")
```
